**Adding Column Week number**

```python
#First lets get a copy of the dataset and get the column alone

date_col = df_accidents.date

#Now lets convert it to a datetime object

date_col = pd.to_datetime(date_col)

date_col
```

Now in order to discretize we will do the following steps:

1) Create a method that takes a column of date objects and returns a column with the respective week of each datetime

2) Add the new column to the dataframe

```python
#Step 1

week = []
for date in date_col:    
    week.append(date.isocalendar()[1])
    
#Step 2

df_accidents["week_number"] = week

df_accidents.head()
```

**Adding New Columns**

we will now add new columns with new insights including:

1) Weekend or not - using the isocalender method where if the weekday is 5/6 then its a weekend

2) Public Holiday or not (Chosen according to http://uk-public-holidays.com/england-and-wales-public-holidays-2016/ )

3) We will also divide the time of the accident into 4 sections:

Morning (05:00-12:00)
Afteroon (12:00-17:00)
Evening (17:00-22:00)
Night (22:00-05:00)

```python
#Weekend (Sat/Sun) or not (1 for true 0 for false)

weekend = []
for date in date_col:    
  if (date.isocalendar()[2] == 5 or date.isocalendar()[2] == 6):
    weekend.append(1)
  else:
    weekend.append(0)

df_accidents["weekend"] = weekend

df_accidents.head()
```

```python
#Public holiday or not (1 for true 0 for false)
from datetime import datetime

newyearsday = datetime.fromisoformat('2016-01-01').date()
goodfriday = datetime.fromisoformat('2016-03-25').date()
eastermonday = datetime.fromisoformat('2016-03-28').date()
earlymay = datetime.fromisoformat('2016-05-02').date()
springbank = datetime.fromisoformat('2016-05-30').date()
summerbank = datetime.fromisoformat('2016-08-29').date()
boxingday = datetime.fromisoformat('2016-12-26').date()
christmas = datetime.fromisoformat('2016-12-27').date()

holiday = []
for date in date_col:    
  if (date.date()== newyearsday or date.date()== goodfriday or date.date()== eastermonday or date.date()== earlymay or date.date()== springbank or date.date()== summerbank or date.date()== boxingday or date.date()== christmas):
    holiday.append(1)
  else:
    holiday.append(0)


df_accidents["public_holiday"] = holiday

df_accidents.head()
```

```python
time_col=df_accidents.time
time_col
```

```python
#Morning (05:00-12:00) ===> 0
#Afteroon (12:00-17:00)===> 1
#Evening (17:00-22:00) ===> 2
#Night (22:00-05:00)   ===> 3


from datetime import time
#time.fromisoformat(df_accidents_copy.time[0]).hour

time_of_day = [] # define array structure
for hour in (df_accidents.time):
    if 5 <= time.fromisoformat(hour).hour < 12:
        time_of_day.append(0)
    elif 12 <= time.fromisoformat(hour).hour < 17:
        time_of_day.append(1)
    elif 17 <= time.fromisoformat(hour).hour < 22:
        time_of_day.append(2)
    else:
        time_of_day.append(3)

df_accidents["time_of_day"] = time_of_day

df_accidents.head()
```

**Encode any categorical feature(s)**

```python
#the idea behind encoding numerical values is: all non numerical  will be encoded numerically (label encoding)
encoding_df_accident = df_accidents.select_dtypes(include=['object']).copy()
#loop to get columns that should be encoded
to_be_encoded=[]
for columnName, columnValue in encoding_df_accident.items():
  count = encoding_df_accident[columnName].unique().size
  to_be_encoded.append(columnName)
#loop on columns that shoud be encoded, make their type as category, then transform this category to numerical value
before_after_encoding=encoding_df_accident.copy()
for item in to_be_encoded:
  encoding_df_accident[item] = encoding_df_accident[item].astype('category')
  encoding_df_accident[item] = encoding_df_accident[item].cat.codes

#create dictionary to map values of columns before and after encoding
encoding_dict=[]
for column in encoding_df_accident:
 encoding_dict.append(dict(zip(encoding_df_accident[column],before_after_encoding[column])))
print(to_be_encoded)
encoding_df_accident.head(25)
encoding_df_accident.dtypes
```

```python
df_accidents = df_accidents.select_dtypes(exclude=['object'])
df_accidents
```

```python
df_copy = df_accidents.copy()
frames = [df_copy, encoding_df_accident]
df_copy = pd.concat(frames, axis=1)
```

```python
df_copy.shape
```

```python
df_accidents=df_copy
```

*We used label encoding instead of one hot encoding to not increase the size of the data frame drastically, since most columns have more than 3 unique values*

```python
#csv file of our encoding
#removing duplicates of dictionaries
unique = []
for dictionary in encoding_dict:
    if dictionary not in unique:
        unique.append(dictionary)
#create dataframe containing the list of unique dictionaries mapping the encoded values
unique_df= pd.DataFrame(unique)
unique_df.head()
#add new dataframe to csv file
unique_df.to_csv("/content/gdrive/My Drive/DataEng/encoding_file_1.csv")
```
